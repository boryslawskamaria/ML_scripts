Visualising the dataset

```python
from sklearn import datasets
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np 
from  sklearn.neural_network import MLPClassifier

l = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
b = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])


iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
xx, yy = np.meshgrid(np.linspace(3.9, 8.1, 300), np.linspace(1.8, 4.6, 300))
a=np.c_[xx.ravel(), yy.ravel()]  
```

No hidden layers 

```python

mlp=MLPClassifier(hidden_layer_sizes=(),activation='logistic',max_iter=5000)
mlp.fit(X, y)
print(mlp.n_layers_)
     
Z = mlp.predict(a)
Z = Z.reshape(xx.shape)
y_=mlp.predict(X)

plt.pcolormesh(xx, yy, Z, cmap=l)
plt.scatter(X[:,0],X[:,1], c=y_,  cmap=b)        
plt.show()
```

2 hidden layers 

```python
mlp=MLPClassifier(hidden_layer_sizes=(2,),activation='logistic',max_iter=5000)
mlp.fit(X, y)
print(mlp.n_layers_)
Z = mlp.predict(a)
Z = Z.reshape(xx.shape)
y_=mlp.predict(X)

plt.pcolormesh(xx, yy, Z, cmap=l)
plt.scatter(X[:,0],X[:,1], c=y_,  cmap=b)        
plt.show()

```

4 hidden layers 

```python

mlp=MLPClassifier(hidden_layer_sizes=(4,),activation='logistic',max_iter=5000)
mlp.fit(X, y)
Z = mlp.predict(a)
Z = Z.reshape(xx.shape)
y_=mlp.predict(X)

plt.pcolormesh(xx, yy, Z, cmap=l)
plt.scatter(X[:,0],X[:,1], c=y_,  cmap=b)        
plt.show()
```

10 hidden layers

```python

mlp=MLPClassifier(hidden_layer_sizes=(10,),activation='logistic',max_iter=5000)
mlp.fit(X, y)
Z = mlp.predict(a)
Z = Z.reshape(xx.shape)
y_=mlp.predict(X)

plt.pcolormesh(xx, yy, Z, cmap=l)
plt.scatter(X[:,0],X[:,1], c=y_,  cmap=b)        
plt.show()
```

100 hidden layers

```python
mlp=MLPClassifier(hidden_layer_sizes=(100,),activation='logistic',max_iter=5000)
mlp.fit(X, y)
Z = mlp.predict(a)
Z = Z.reshape(xx.shape)
y_=mlp.predict(X)

plt.pcolormesh(xx, yy, Z, cmap=l)
plt.scatter(X[:,0],X[:,1], c=y_,  cmap=b)        
plt.show()

mlp=MLPClassifier(hidden_layer_sizes=(100,100,100),activation='logistic',max_iter=5000000)
mlp.fit(X, y)
Z = mlp.predict(a)
Z = Z.reshape(xx.shape)
y_=mlp.predict(X)

plt.pcolormesh(xx, yy, Z, cmap=l)
plt.scatter(X[:,0],X[:,1], c=y_,  cmap=b)        
plt.show()
```
