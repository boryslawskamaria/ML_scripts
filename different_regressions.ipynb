Linear and polynomial regressions of degree 3 and 17 were compared and pictured on the graph below.

---

 Standard ridge regression was also calculated with the lambda parameter of 1 and 10, taking into account  all the coefficients in the penalty function. The new curves were compared with 17 degree polynomial regression. 


```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.linear_model as lm
#form sklearn.linear_model import LinearRegression
#form sklearn.linear_model import Ridge # do ridge regression
import sklearn.preprocessing as pr
# do wyrenerowania danych X w dopasowaniu liniowym (kolejnych poteg zmiennej x)
#form sklearn.preprocessing import PolynomialFeatures

x=np.linspace(0,2*np.pi,40)
y=np.sin(x)+np.random.normal(0,0.4,40)
plt.scatter(x,y)

x=x[:, np.newaxis]
r1=lm.LinearRegression()
r1.fit(x,y)
y_=r1.predict(x)
plt.plot(x,y_)

poly=pr.PolynomialFeatures(3)
r1.fit(poly.fit_transform(x),y)
y_=r1.predict(poly.fit_transform(x))
plt.plot(x,y_)

poly=pr.PolynomialFeatures(17)
r1.fit(poly.fit_transform(x),y)
y17=r1.predict(poly.fit_transform(x))
plt.plot(x,y17)

plt.show()

r=lm.Ridge(alpha=1 )
r.fit(x,y)
y_=r.predict(x)
l1=(r.intercept_, r.coef_)
plt.plot(x,y_)

r=lm.Ridge(alpha=10)
r.fit(x,y)
y_=r.predict(x)
l10=(r.intercept_, r.coef_)
plt.plot(x,y_)

r=lm.Ridge(alpha=10, fit_intercept=False)
r.fit(x,y)
y_=r.predict(x)
l10=(r.intercept_, r.coef_)
plt.plot(x,y_)

p=pr.PolynomialFeatures(1)
r=lm.Ridge(alpha=1)
r.fit(p.fit_transform(x),y)
y_=r.predict(p.fit_transform(x))
plt.plot(x,y_)

plt.plot(x,y17)
plt.scatter(x,y)
plt.show()
```

Using the normal equation for linear regression with regularization ($ X '^ TX' + lambda I_ {p-1} beta '= X' ^ Ty $  and  $ beta_0 $ equals the mean of $ y $) the beta for the data from the previous example was calculated and compared with the results obtained earlier.

```python
l=1
a=np.dot(x.T,x)+l*np.identity(1)

b0=np.mean(y)
b1=np.dot(np.dot(x.T,np.array(y, ndmin=2).T),np.linalg.inv(a))
print("l1" , l1, "b0", b0, "b1", b1)

l=10
a=np.dot(x.T,x)+l*np.identity(1)
b0=np.mean(y)
b1=np.dot(np.dot(x.T,np.array(y, ndmin=2).T),np.linalg.inv(a))
print("l10", l10, "b0", b0, "b1", b1)
```
