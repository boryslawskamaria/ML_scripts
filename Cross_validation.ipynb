A random straight line passing through the center of the coordinate system and classification of the data 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
N=1000

df1 = 2*pd.DataFrame(np.random.randn(N, 2), columns=['A', 'B'])
df2 = 20+10*pd.DataFrame(np.random.randn(N, 2), columns=['A', 'B'])

Data=df1.append(df2)


s='b'*N+'r'*N
plt.scatter(Data['A'],Data['B'], s=5,alpha=1,c=list(s))
plt.show()

X=Data.values
from sklearn.model_selection import train_test_split
y=[x == "b" for x in s]
model1 = linear_model.LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True)
model2 = linear_model.LinearRegression(copy_X=True, fit_intercept=True, n_jobs=1, normalize=False)


X_t, X_test,y_t,y_test = train_test_split(Data, y)

model1.fit(X_t,y_t)
model2.fit(X_t,y_t)
def model1_pred(x,y):
    v=np.array([x,y]).reshape(1,-1)

    return model1.predict(v)
model1_pred=np.vectorize(model1_pred)
def model2_pred(x,y):
    v=np.array([x,y]).reshape(1,-1)

    return model2.predict(v)>0.5

model2_pred=np.vectorize(model2_pred)
x = y = np.arange(-20, 60, 0.1)
x, y = np.meshgrid(x, y)
z=model2_pred(x,y)

plt.contourf(x,y,z)
plt.scatter(Data['A'],Data['B'], s=5,alpha=1,c=list(s))
plt.show()
x = y = np.arange(-20, 60, 0.1)
x, y = np.meshgrid(x, y)
z=model1_pred(x,y)

plt.contourf(x,y,z)
plt.scatter(Data['A'],Data['B'], s=5,alpha=1,c=list(s))
plt.show()
```

Logistic regression on the Iris dataset

```python
from sklearn import datasets
from sklearn import linear_model as lm, datasets
import matplotlib.pyplot as plt
import numpy as np
from sklearn import  linear_model

iris = datasets.load_iris()

x = iris.data[:, :2]  
y = iris.target[:]
plt.scatter(x[:, 0], x[:, 1], c=y)
plt.show()

iris=lm.LogisticRegression().fit(x, y)

xmin, xmax = x[:, 0].min(), x[:, 0].max()
ymin, ymax = x[:, 1].min(), x[:, 1].max()


xx, yy = np.meshgrid(np.arange(xmin, xmax, 0.01), np.arange(ymin, ymax, 0.01))



z = iris.predict(np.c_[xx.ravel(), yy.ravel()])

z = z.reshape(xx.shape)

cmap = plt.cm.get_cmap("gist_rainbow")
cmap.set_under("red")
cmap.set_over("yellow")
plt.figure()
plt.pcolormesh(xx, yy, z, cmap=cmap,  vmin=None, vmax=None, shading='flat')

plt.scatter(x[:, 0], x[:, 1], c=y, marker='*')

plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()

from sklearn.preprocessing import PolynomialFeatures

from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(linear_model.LogisticRegression(),param_grid={'C': [0.1,0.5,1,10,100]}, cv=2, refit=False)
print(grid.fit(x, y))
print(grid.best_params_)
```

Logistic and linear regression using the below dataset:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model

N=1000

df1 = 2*pd.DataFrame(np.random.randn(N, 2), columns=['A', 'B'])
df2 = 20+10*pd.DataFrame(np.random.randn(N, 2), columns=['A', 'B'])


Data=df1.append(df2)
Data['C']=0
Data['C'][range(0,N)]=1

s='b'*N+'r'*N
plt.scatter(Data['A'],Data['B'], s=5, alpha=1, c=list(s))

```

```python

R=linear_model.LogisticRegression(C=1000)
R.fit(Data[['A','B']],Data['C'])
print(R.coef_, R.intercept_)


S=linear_model.LinearRegression()
S.fit(Data[['A','B']],Data['C'])
print(S.coef_, S.intercept_)


b=-1*R.coef_[0][1]
a=R.coef_[0][0]
c=R.intercept_[0]


x=range(-15,30)
line=a/b*x+c/b
plt.plot(x, line, 'black')

b=-1*S.coef_[1]
a=S.coef_[0]
c=S.intercept_

x=range(-10,60)
line=a/b*x+c/b
plt.plot(x, line, 'black')
plt.show()
```

 "True/False Positive/Negative" table and sensitivity, specificity, positive predictive value and F1 value calculation

```python

linregression=S.predict(Data[['A','B']])
X1=np.bincount(linregression[0:1000]>0.5)
X2=np.bincount(linregression[1000:2000]<0.5)
TP=X1[1]
FP=X2[1]
FN=X1[0]
TN=X2[0]
print (TN, TP, FN, TN)


sensitivity=TP/(TP+FN)
ppv=TP/(TP+TN)
specificity=FP/(TN+FP)
f1= 2*sensitivity*ppv/(sensitivity+ppv)

print(sensitivity, ppv, specificity, f1)

logregression=R.predict(Data[['A','B']])
X1=np.bincount(logregression[0:N])
X2=np.bincount(logregression[N+1:2*N])
TP=X1[1]
FP=X2[1]
FN=X1[0]
TN=X2[0]
print (TN, TP, FN, TN)


sensitivity=TP/(TP+FN)
dwp=TP/(TP+TN)
specificity=FP/(TN+FP)
f1= 2*sensitivity*ppv/(sensitivity+ppv)

print(sensitivity, ppv, specificity, f1)

```
