The gradient descent algorithm is written to find the global minimum.

---


Its functionality is presented based on the example function $ f (x) = x ^ 2 $, starting from  $ x = 2 $. The first 5 iterations are presented on the graph.

```python
import matplotlib.pyplot as plt
import numpy as np

alfa=0.3
n=100
x=np.ones(n)

x[0]=2
for i in range(0,n-1):
    x[i+1]=x[i]*(1-alfa*2)


y=np.arange(-6, 6, 0.1)
a=y**2

plt.plot(y, a, linewidth=2.0,linestyle='--')
plt.plot(x[0:5], x[0:5]**2, 'ro')
```

Its functionality is presented based on the example function $f(x,y)=x^2-2x+y^2$,  starting from  $(x,y)=(2,2)$. The first 5 iterations are presented on the graph.

```python
n=50
alfa=0.3
x=np.ones((n,2))
x[0]=[2,2]

for i in range(0,n-1):
    x[i+1][0]=x[i][0]-alfa*(2*x[i][0]-2)
    x[i+1][1]=x[i][1]-alfa*2*x[i][1]


xx= np.arange(-2.0, 5.0, 0.01)
yy= np.arange(-3.0, 4.0, 0.01)
a, b = np.meshgrid(xx, yy)

c= a**2-2*a+b**2
plt.figure()
plt.contour(a, b, c)
plt.plot(x[:,0][0:5],x[:,1][0:5],'ro')
```

Its functionality is presented based on the example function $f(x)=x^2-y^2$,  starting from  $x=5$ and $x=1$. The first 5 iterations are presented on the graph.

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

 
alfa=0.3
n=100
x=np.ones((n,2))
x[0]=[5,1]
for i in range(1,n-1):
    x[i+1][0]=x[i][0]-alfa*2*x[i][0]
    x[i+1][1]=x[i][1]-alfa*(-2)*x[i][1]


xx= np.arange(-7.0, 7.0, 0.02)
yy= np.arange(-7.0, 7.0, 0.02)
a, b = np.meshgrid(xx, yy)
c=a**2-b**2
plt.figure()
plt.contour(a, b, c)
plt.plot(x[:,0][0:5],x[:,1][0:5],'ro')

```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(a, b, c, alpha=0.3)
ax.scatter(x[:,0][0:5],x[:,1][0:5], color='r')

plt.show()
```
