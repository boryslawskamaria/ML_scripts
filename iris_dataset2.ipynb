Based on the Iris dataset,  a logistic regression model was trained, where the independent variable is column D. 

1.   "True / False,  Positive / Negative" table was presented
2.   Sensitivity, specificity, positive predictive value and F1 value were calculated


The second model dataset is the same dataset, but with the square root of the independent variable. 

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets, linear_model

iris = datasets.load_iris()

x = iris.data[iris.target!=0]
y = iris.target[iris.target!=0]
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)
ax.scatter(x[:, 0], x[:, 1], x[:, 2], c=y, cmap=plt.cm.Set1, edgecolor='k', s=40)


A=linear_model.LogisticRegression(C=10000)
A.fit(x[:,0:3], y)
p=A.predict(x[:,0:3])
print(A.coef_, A.intercept_)
xx=np.arange(4.5,8.5,0.1)
yy=np.arange(1,5,0.1)
xx, yy = np.meshgrid(xx, yy)
c=A.coef_[0][2]
z=-xx*A.coef_[0][0]/c-yy*A.coef_[0][1]/c-A.intercept_[0]/c
ax.plot_surface(xx, yy, z, alpha=0.4, color='grey')
plt.show()

X1=np.bincount(p[0:50]==1)
X2=np.bincount(p[50:100]==2)
TP=X1[1]
FP=X2[1]
FN=X1[0]
TN=X2[0]
print ("TN" , TN, "TP", TP, "FN", FN, "TN", TN)


sensitivity=TP/(TP+FN)
ppv=TP/(TP+TN)
specificity=FP/(TN+FP)
f1= 2*sensitivity*ppv/(sensitivity+ppv)


print("sensitivity", sensitivity,"ppv" , ppv, "specificity" , specificity, "f1", f1)

```
