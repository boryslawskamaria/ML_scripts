The logistic model with and without regularization with differently selected parameters is trained here. Each model is illustrted with an appropriate graph and compared with the basic characteristics of the models (using the "True / False Positive / Negative" table). 

```python
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression as LR
import numpy as np

iris = datasets.load_iris()
X = iris.data[:, :2]  
y = iris.target
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()

a=LR()
a.fit(X, y)
y_=a.predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_)
plt.show()


X1=np.bincount(y_[0:50]==0)
X2=np.bincount(y_[50:100]==1)
print("Regression 1:","TP: ", X1[1],"FN: ", X1[0], "FP: ", X2[1], "TN: ", X2[0])
a=X1[1]/(X1[1]+X1[0])
b=X1[1]/(X1[1]+X2[0])
print("sensitivity:", a, " specificity:", X2[1]/(X2[0]+X2[1]), " PPV:", b, "F1:", 2*a*b/(a+b))
```

```python
Logistic regression is trained using optimal parameters and using cross-validation.
```

```python
from sklearn.linear_model import LogisticRegressionCV as LRCV
from sklearn.linear_model import LogisticRegression as LR
from sklearn.linear_model import RidgeClassifier as RC
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report


iris = datasets.load_iris()
X = iris.data[:, :2]  
y = iris.target
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()

a=LRCV()
a.fit(X[:,0:2], y)
y_=a.predict(X[:,0:2])
plt.scatter(X[:,0],X[:,1],c=y_)
plt.show()
print(classification_report(y,y_))


#6.4
a=LR()
a.fit(X[:,0:2], y)
y_=a.predict(X[:,0:2])
plt.scatter(X[:,0],X[:,1], c=y_)
plt.show()
print(classification_report(y,y_))

a=RC(alpha=5)
a.fit(X[:,0:2], y)
y_=a.predict(X[:,0:2])
plt.scatter(X[:,0],X[:,1], c=y_)
plt.show()
print(classification_report(y,y_))

a=RC(alpha=30)
a.fit(X[:,0:2], y)
y_=a.predict(X[:,0:2])
plt.scatter(X[:,0],X[:,1], c=y_)
plt.show()
print(classification_report(y,y_))

a=RC(alpha=0.1)
a.fit(X[:,0:2], y)
y_=a.predict(X[:,0:2])
plt.scatter(X[:,0],X[:,1], c=y_)
plt.show()
print(classification_report(y,y_))
```
